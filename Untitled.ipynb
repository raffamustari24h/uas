```python
import pandas as pd
import matplotlib.pyplot as plt

file_path =r"C:\Users\My Acer\Documents\uas dasprog raffa\hospital data analysis.csv"

df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
plt.hist(df['Cost'], bins=20, color='green', edgecolor='black')
plt.title('Distribusi Biaya Perawatan')
plt.xlabel('Biaya (Cost)')
plt.ylabel('Jumlah Pasien')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\My Acer\Documents\uas dasprog raffa\hospital data analysis.csv"
df = pd.read_csv(file_path)

avg_stay = df.groupby('Condition')['Length_of_Stay'].mean().sort_values()

plt.figure(figsize=(12, 6))
avg_stay.plot(kind='barh', color='pink')
plt.title('Rata-rata Lama Rawat Inap per Kondisi')
plt.xlabel('Rata-rata Lama Rawat Inap (hari)')
plt.ylabel('Kondisi')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```
